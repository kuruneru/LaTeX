# <mark>Python, IPython, Jupyter Notebook, Google Colaboratory</mark>
Python は，プログラム言語の1つで，インタープリタ型言語です．インタープリタ型言語は，プログラムのソースコードをそのままコンパイルすることなしに実行します．

インタープリタのため，コードの文字列を入力してすぐ実行することができるため，対話型（インタラクティブ）にプログラムができ，コマンドラインのように1行1行実行できるようにした対話型 python を interactive python (IPython) と言います．

Jupyter Notebookとはノートブックと呼ばれる単位でpython等のプログラミングやMarkdownによる記述ができるツールです．IPython をブラウザから実行できるようにしたものです．Chrome, firefox, Edgeなどブラウザさえあればどこでもプログラミングが出来ます．また，プログラムとその実行結果とその記録を簡単に作成できるので，後から振り返る際にも便利です．

Google colaboratory は，これをクラウドサービス化したものです．

## 画面出力 print
### 文字や処理結果などを表示させる

```python
string = 'Hello Exp2! '
number = 2025

#変数の中身をそのまま表示
print(string)
print(number)

#format関数を利用し，文字列と数値を結合
print('Welcome to exp2 {0}'.format(number))

#+演算子で文字列と変数を組み合わせることができる（連結）
print(string + '2025')

#+演算子で連結可能なのは文字列だけなので，数値などは文字列に変換してから連結する
print(string + str(number))

#formatと連結の組み合わせでももちろん大丈夫
print(string +'{0}'.format(number))

#C言語でおなじみの%記法もある
print('%s%d' % (string, number))

#引数endに''を指定すれば出力の後改行されない（複数のprintの結果を一行に）
print(string, end='')
print(number)
```

## forを使った繰り返し制御
### pythonにおけるfor文の流れは原則「データの集まり」から「データを一つずつ取り出す」作業

    for 変数 in データの集まり:
        処理

```python
#文字の集まりである文字列から一つずつ取り出す
for char in string:
    print(char)

#指定回数繰り返す
for i in range(5):
    print(i)

#区間を指定したrangeも可能
for i in range(1, 6):
    print(i)
```

## 条件分岐 if
### 条件式に合わせ処理を分岐させる
    if 条件式1:
        条件式1が真の時に実行する処理
    elif 条件式2:
        条件式1が偽，条件式2が真の時に実行する処理
    else:
        全ての条件式が偽の時に実行する処理
        
### 条件式
#### 比較演算子
    演算子  使用例    意味
    ==      a == b   bがaに等しい
    !=      a != b   bがaに等しくない
    >       a > b    bよりaが大きい
    >=      a >= b   bよりaが大きいか等しい
    <       a < b    bよりaが小さい
    <=      a <= b   bよりaが小さいか等しい
#### 論理演算子
    演算子  使用例    意味
    and     a and b  aとbが共に真の場合に真
    or      a or b   aかbの少なくとも1つが真の場合に真
    not     a not b  aが真の時に偽、偽の時に真


```python
#2025年10月の1日～31日の各日が何の日か調べる
for d in range(1,32):
    print(str(d) + ': ', end='')
    if d == 13:
        print('スポーツの日')
    elif d == 18:
        print('大学祭')
    elif d == 19:
        print('大学祭')
    elif d % 7 == 4:
        print('土曜日')
    elif d % 7 == 5:
        print('日曜日')
    else:
        print('平日')
```

## 関数
### 様々な処理をあらかじめ定義できる．
定義前に呼び出す前方参照はpythonでは許されていない．（必ずdefで定義した後に呼び出す）

    def 関数名(引数)
        処理内容
        return（戻り値）#省略時はNoneが返る

```python
#条件分岐の例を関数化
def nameOfDay(d):
    if d == 13:
      name = 'スポーツの日'
    elif d == 18:
      name = '大学祭'
    elif d == 19:
      name = '大学祭'
    elif d % 7 == 4:
      name = '土曜日'
    elif d % 7 == 5:
      name = '日曜日'
    else:
      name = '平日'
    return name

for d in range(1,32):
    print(str(d) + ': ' + nameOfDay(d))
```
